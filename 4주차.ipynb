```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# column vector : input x, output y
x = np.array([0.2, 0.4, 0.7, 1.2, 1.5, 1.7, 2.0, 2.5, 2.7, 3.0, 3.4, 3.8, 4.3, 4.4, 4.9]).reshape(-1, 1)
y = np.array([0.6, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.2, 2.6, 2.7, 3.0, 3.0, 3.5, 3.7, 3.8]).reshape(-1, 1)

# A 만들기
m = y.shape[0]
A = np.hstack([np.ones([m, 1]), x])

# 다른 방법: 0제곱으로 1 만들기
A = np.hstack([x**0, x])

# 매트릭스로 변환
A = np.asmatrix(A)

# w 계산 [원점 높이], [기울기] = [0.64805086], [0.66973398]
w = (A.T * A).I * A.T * y
print(w)

#x와 y 자표 설정
xp = np.arange(0, 5, 0.01).reshape(-1, 1)
yp = w[0, 0] + w[1, 0] * xp

plt.figure(figsize = (6, 4))

plt.title('Training Data', fontsize = 16)
plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.axis('equal')
plt.grid(alpha = 0.4)

plt.plot(x, y, 'bo')
plt.plot(xp, yp, 'r')

plt.show()
```
